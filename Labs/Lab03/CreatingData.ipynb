```python
%matplotlib inline
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import random
```

```python
# Create somewhat clustered random data
X, y_true = make_blobs(n_samples = 300, centers = 4, 
              cluster_std = 0.6, random_state = 0)
```

```python
# Check datatype
type(y_true)
```

```python
# Use the original labels as the "class year"
year = y_true 
```

```python
# Attempting various plots of the data output
plt.scatter(-1*X[:,0],X[:,1], s=50)
```

```python
# Convert the first variable to "gym time"
gym_time = -1*X[:,0]
mx = np.max(gym_time)
mn = np.min(gym_time)
spread = (mx - mn)
gym_time = 150*((gym_time - mn)/spread) 
gym_time = np.around(gym_time)
```

```python
# Convert the first variable to "coffee"
coffee = X[:,1]
mx = np.max(coffee)
mn = np.min(coffee)
spread = (mx - mn)
coffee = 2.5*((coffee - mn)/spread) 
coffee = np.around(coffee, decimals = 1)
```

```python
# Use a plot to check the conversions

plt.scatter(gym_time,coffee, s=50)
```

```python
# Create more data, that is again clustered

Y, z_true = make_blobs(n_samples = 300, centers = 3, 
              cluster_std = 1, random_state = 0)
```

```python
# Attempting various plots of the data output
plt.scatter(Y[:,0],Y[:,1], s=50)
```

```python
# Converting the second variable to "GPA"
gpa = Y[:,1]
mx = np.max(gpa)
mn = np.min(gpa)
spread = (mx - mn)
gpa = 2*((gpa - mn)/spread)+2
gpa = np.around(gpa, decimals = 2)
```

```python
# Use a plot to check the conversions
plt.scatter(gpa,Y[:,1], s=50)
```

```python
# Converting the first variable to represent "hours slept"
sleep = Y[:,0]
mx = np.max(sleep)
mn = np.min(sleep)
spread = (mx - mn)
sleep = 5.5*((sleep - mn)/spread)+3.5
sleep = np.around(sleep, decimals = 2)
```

```python
# Use a plot to check the conversions
plt.scatter(sleep,gpa, s=50)
```

```python
# Creating an off-balance of favorite colors
colors1 = ["red","blue","yellow","purple", "green"]
fav_color = random.choices(colors1, k=100)
colors2 = ["red","blue","yellow","purple", "green", "pink", "grey", "tangerine"]
fav_color2 = random.choices(colors2, k=200)
fav_color.extend(fav_color2)
print(fav_color)
```

```python
# creating a list of birth months
mon = ["jan", "feb","mar","apr","may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
month = random.choices(mon, k=300)
```

```python
# Converting numerical variables from an array to a list
coffee = list(coffee)
gpa = list(gpa)
sleep = list(sleep)
gym_time = list(gym_time)
```

```python
# Following https://www.ritchieng.com/creating-dataframe-from-objects/
# for creating a pandas data frame

student_info = {
    'coffee': coffee,
    'month': month,
    'sleep': sleep,
    'gym':gym_time,
    'color': fav_color,
    'gpa':gpa
}

cols = ['coffee','month','sleep','gym','color','gpa']
```

```python
# Creating the index for the data frame
students = []
for i in range(1,301):
    name = "student_"+str(i)
    students.append(name)
```

```python
# Making the data frame with the listed columns and the student numbers as the row names
students = pd.DataFrame(student_info, columns = cols, index = students)
```
