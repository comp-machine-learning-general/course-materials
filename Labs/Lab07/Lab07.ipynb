In supervised machine learning, we have extra information about some of our observations that is regarded as separate from the variables. This extra information could be a rating, a label, a grouping, or an outcome, among other things. Colloquially, in this course, we might refer to this extra information as _answers_. (**Note** - Knowing that the data falls into two groups is not the supervision that we are referring to.)

However the general process that we have done is known as _k-nearest neighbor_ or kNN, for short.

## k-Nearest Neighbors

To apply kNN, we need to have a dataset with a few labeled points. Then we need to decide how many labeled neighbors we want each data point to "consult" to determine the labeling. This is the $k$ in kNN.
